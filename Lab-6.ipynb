```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('dataset.csv')

# Create a scatter plot
plt.scatter(data['age'], data['avg_glucose_level'], c=data['heart_disease'], cmap='viridis', alpha=0.5)

# Add color bar to indicate heart disease (if it's binary)
plt.colorbar(label='Heart Disease')

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.title('Scatter Plot of Age vs. Average Glucose Level')

# Show the plot
plt.show()
```
